# Sequence convergence

1. Show with the definition that the sequence $\left\{ \dfrac{1}{k} \right\}_{k=1}^\infty$ does not have a linear convergence rate (but it converges to zero).\
  **Answer:** $$\lim\limits_{k\rightarrow\infty}\dfrac{1}{k} = 0 \Rightarrow r_k = \dfrac{1}{k} - 0 = \dfrac{1}{k}$$
  Suppose $r_k$ has linear convergence, then exists $0 < q < 1$ and $0 < C <\infty$ such that:
  $$\dfrac{1}{k}\leq Cq^k \Rightarrow \forall k: C^{-1}\leq kq^k \Rightarrow C^{-1}\leq \lim\limits_{k\rightarrow\infty}kq^k =0$$
We got a contradiction, so the convergence is not linear

  $$ r_{k} = \dfrac{1}{k^2} \leq 1\cdot k^{-2}$$

1. Show with the definition that the sequence $\left\{ \dfrac{1}{k^k} \right\}_{k=1}^\infty$ does not have a quadratic convergence rate (but it converges to zero).\
  **Answer:**
  $$\lim\limits_{k\rightarrow\infty}\dfrac{1}{k^k} = 0 \Rightarrow r_k = \dfrac{1}{k^k} - 0 = \dfrac{1}{k^k}$$
  Suppose  has quadratic convergence, then exists $0 < q < 1$ and $0 < C < \infty$ such that:
  $$\dfrac{1}{k^k}\leq Cq^{2^k} \Rightarrow \forall k: C^{-1}\leq k^kq^{2^k} \Rightarrow C^{-1}\leq \lim\limits_{k\rightarrow\infty}k^kq^{2^k} =0$$
  We got a contradiction, so the convergence is not quadratic.\
  *Limit calculation:*
  $$\lim\limits_{k\rightarrow\infty}k^kq^{2^k} = \lim\limits_{k\rightarrow\infty}\exp(k\ln k + 2^k\ln q)$$
  Since $0 < q < 1 \Rightarrow \ln q < 0$ and $2^k$ is growing much faster than $k \ln k$ we get:
  $$ k\ln k + 2^k\ln q \xrightarrow{k\rightarrow\infty} -\infty \Rightarrow \lim\limits_{k\rightarrow\infty}\exp(k\ln k + 2^k\ln q) = 0$$

1. Determine the convergence or divergence of a given sequence $r_{k} = 0.707^k$.\

  **Answer:**\

1. Consider the function $$f(x) = (x + \sin x) e^x, \;\;\; x \in [-20, 0]$$ 
    ![](https://raw.githubusercontent.com/MerkulovDaniil/optim/master/docs/exercises/Unimodal.svg)

    Consider the following modification of solution localization method, in which the interval $[a,b]$ is divided into $2$ parts in a fixed proportion of $t: x_t = a + t*(b-a)$ (maximum twice at iteration - as in the dichotomy method). Experiment with different values of $t \in [0,1]$ and plot the dependence of $N (t)$ - the number of iterations needed to achieve $\varepsilon$ - accuracy from the $t$ parameter. Consider $\varepsilon = 10^{-7}$. Note that with $t = 0.5$ this method is exactly the same as the dichotomy method.

```
import math
import numpy as np
import matplotlib.pyplot as plt

def f(x):
  return (x+ math.sin(x)) * math.exp(x)

def line_search(a, b, t, epsilon):
  N = 0
  c = a + t * (b-a)
  while abs(b-a) > epsilon:
    N += 1
    y = a + t * (c-a)
    if f(y) <= f(c):
      b = c
      c = y
    else:
      z = c + t * (b-c)
      if f(c) <= f(z):
        a = y
        b = z
      else:
          a = c
          c = z
  return N

def graph(N):
  t = np.arange(0+1/N, 1-1/N, 1/(N+1))
  y = [0] * (N-1)
  for i in range(1, N-1):
    y[i] = line_search(-20, 0, i/N, 10**(-7))
  fig = plt.figure(figsize=(9,6))
  plt.plot(t, y)
  plt.xlabel('t')
  plt.ylabel('N')
  plt.title('N(t)')
  plt.figure(figsize=(30,20))
  plt.show()

graph(100)
```

3. Show that the one-dimensional minimizer of a strongly convex quadratic function always satisfies the Goldstein conditions.\
  **Answer:**\
  Consider $\mathbf{A}\in \mathbb{S}^n_{++}$ (else let's take $\mathbf{A}' = \frac{\mathbf{A}^\mathrm{T} +\mathbf{A}}{2}$) and the strongly convex quadratic function:
  $$ f(x) = \dfrac{1}{2}x^\mathrm{T}\mathbf{A}x + b^\mathrm{T}x+c$$
  $$ \nabla f(x) = \mathbf{A}x + b, \hspace{0.7cm} x_{k+1} = x_k - \eta_k \nabla f(x_k)$$
  The value of $\eta_k$ was found at the seminar ($\eta_k = \text{arg}\min\limits_{k\in \mathbf{R}^+}f(x_{k+1}))$:
  \
  $$ \eta_k = \dfrac{\nabla f(x_k)^\mathrm{T}\nabla f(x_k)}{\nabla f(x_k)^\mathrm{T}\mathbf{A}\nabla f(x_k)} = \dfrac{(x_k^\mathrm{T}\mathbf{A} + b^\mathrm{T})\nabla f(x_k)}{\nabla f(x_k)^\mathrm{T}\mathbf{A}\nabla f(x_k)} $$
  The Goldstein conditions (denote $\nabla f(x_k) = \nabla f_k$):
  $$f(x_k) - \rho\eta_k\nabla f_k^\mathrm{T}\nabla f_k \leq f(x_k +\eta_kd_k) \leq $$ $$\leq f(x_k) - (1-\rho)\eta_k\nabla f_k^\mathrm{T}\nabla f_k \hspace{0.7cm} \forall \rho \in(0.5; 1)$$

  Then the first part of the inequality takes the form:
  $$f(x_k) -\rho\eta_k\nabla f_k^\mathrm{T}\nabla f_k \leq f(x_k) -\dfrac{1}{2}\eta_k\nabla f_k^\mathrm{T}\nabla f_k$$
  $$\Rightarrow  \rho \geq \dfrac{1}{2}$$

Implement Rastrigin function $f: \mathbb{R}^d \to \mathbb{R}$ for d = 10. [link](https://www.sfu.ca/~ssurjano/rastr.html)

$$
f(\mathbf{x})=10 d+\sum_{i=1}^{d}\left[x_{i}^{2}-10 \cos \left(2 \pi x_{i}\right)\right]
$$

* Consider global optimization from [here](https://docs.scipy.org/doc/scipy/reference/optimize.html#global-optimization).
* Plot 5 graphs for different $d$ from {2,4,8,16,32}. On each graph you are to plot $f$ from $N_{fev}$ for 5 methods: `basinhopping`, `brute`, `differential_evolution`, `shgo`, `dual_annealing` from scipy, where $N_{fev}$ - the number of function evaluations. This information is usually avalable from `specific_optimizer.nfev`. If you will need bounds for the optimizer, use $x_i \in [-5, 5]$.

```
import scipy
from scipy import optimize
import random
def rastr(x): #The Rastrigin function
    return np.sum(1 + x*x-10*np.cos(2 * np.pi * x), axis=0)
def basinhopping():
  x = np.array([2,4,8,16,32])
  y = [0]*5
  for j in range(5):
    d = x[j]
    x0 = [0]*d
    for i in range(0, d):
      x0[i] = random.random() * 10 -5
    mybounds=[[-5,5] for i in range(d)]
    y[j] = scipy.optimize.basinhopping(rastr, np.array(x0), minimizer_kwargs={"bounds":mybounds}).nfev
  fig = plt.figure(figsize=(9,6))
  plt.title('basinhopping')
  plt.plot(x,y)
  plt.show()
  print(y)

basinhopping()
```

```
def brute():
  x = np.array([2,4,8])
  fig = plt.figure(figsize=(9,6))
  plt.title('brute')
  plt.plot(x, 20**x)
  plt.show()
brute()
```

```
def differential_evolution():
  x = np.array([2,4,8,16,32])
  y = [0]*x.size
  for j in range(x.size):
    d = x[j]
    x0 = [0]*d
    for i in range(0, d):
      x0[i] = random.random() * 10 -5
    mybounds=[[-5,5] for i in range(d)]
    y[j] = scipy.optimize.differential_evolution(rastr, bounds=mybounds).nfev
  fig = plt.figure(figsize=(9,6))
  plt.title('differential evolution')
  plt.plot(x,y)
  plt.show()
  print(y)
differential_evolution()
```

```
def shgo():
  x = np.array([2,4,8])
  y = [0]*x.size
  for j in range(x.size):
    d = x[j]
    x0 = [0]*d
    for i in range(0, d):
      x0[i] = random.random() * 10 -5
    mybounds=[[-5,5] for i in range(d)]
    y[j] = scipy.optimize.shgo(rastr, bounds=mybounds).nfev
  fig = plt.figure(figsize=(9,6))
  plt.title('shgo')
  plt.plot(x,y)
  plt.show()
  print(y)
shgo()
```

Let the price of the cable to be pulled from the station to each house independently be determined by some function p(d) . $$
P(w, x) = \sum\limits_{i=1}^N p(d_i) = \sum\limits_{i=1}^N p(|w - x_i|)

$w$ - station location, $x_i$ - location of $i$ house.

Write analytical solution $w^*$ for minimization $P(w,x)$, if $p(d) = d^2$

**Answer:**\
$$ P(w, x) =\sum\limits_{i=1}^N p(d_i) = \sum\limits_{i=1}^N |w - x_i|^2 = N \cdot w^2 - 2 \cdot\left(\sum\limits_{i=1}^Nx_i\right)w+ \sum\limits_{i=1}^N x_i^2$$

The objective function value is
$$ P(w_0, x) = c - \dfrac{b^2}{4a} = \sum\limits_{i=1}^N x_i^2 - \dfrac{\left(\sum\limits_{i=1}^Nx_i\right)^2}{N}$$

Write function `gradient_descent`, which returns $w_k$ after a fixed number of steps.   

$$
w_{k+1} = w_k - \mu \nabla P(w_k)
$$

```
def gradient_descent(x, dP, w0, mu, Nsteps):
    k = 0
    w_k = w0
    while k < Nsteps:
      w_k = w_k - mu * dP(w_k, x)
      k +=1
    return w_k

gradient_descent(x, dP, 0.5, 1, 1)
```

Write function `newton_descent`, which return all optimization trajectory. Update rule:

$$
w_{i+1} = w_{i} - \nabla^{2} f\left(w_{i}\right)^{-1} \nabla f\left(w_{i}\right)
$$

```
def newton_descent(x, dP, ddP, w0, Nsteps):
    trajectory = [0] * (Nsteps +2)
    trajectory[0] = w0
    i = 1
    while i < (Nsteps + 1):
      trajectory[i+1] = trajectory[i] - dP(trajectory[i], x)/ddP(trajectory[i], x)
      i+=1
    return trajectory
```

## Functional class
A function is said to belong to the class $f \in C^{k,p}_L (Q)$ if it $k$ times is continuously differentiable on $Q$ and the $p$th derivative has a Lipschitz constant $L$. 

$$
\||\nabla^p f(x) - \nabla^p f(y)\| \leq L \|x-y\|, \qquad \forall x,y \in Q
$$

The most commonly used $C_L^{1,1}, C_L^{2,2}$ on $\mathbb{R}^n$. 
Note that:
* $p \leq k$
* If $q \geq k$, then $C_L^{q,p} \subseteq C_L^{k,p}$. The higher the order of the derivative, the stronger the constraint (fewer functions belong to the class)

Prove that the function belongs to the class $C_L^{2,1} \subseteq C_L^{1,1}$ if and only if $\forall x \in \mathbb{R}^n$:

$$
\|\nabla^2 f(x)\| \leq L
$$

Prove also that the last condition can be rewritten, without generality restriction, as follows:

$$
-L I_n \preceq \nabla^2 f(x) \preceq L I_n
$$

Note: by default the Euclidean norm is used for vectors and the spectral norm is used for matrices.

**Answer:**
$$ f(x) \in C_L^{2,1} \Leftrightarrow \|f(x+\Delta x) - f(x)\| \leq L\|\Delta x\| \Leftrightarrow $$
$$ \|\nabla^2f(x)\|\leq L \hspace{0.5cm} \forall x \in \mathbb{R}^n, \text{Q.E.D..}$$
And also
$$ \|\nabla^2f(x)\|\leq L \Leftrightarrow -L \leq \lambda(\nabla^2f(x)) \leq L \Leftrightarrow$$
$$\forall x, y: \ -Ly \leq \nabla^2f(x) y \leq Ly \Leftrightarrow -L I_n \preceq \nabla^2 f(x) \preceq L I_n$$

For example, steepest descent algorithm will be plotted with the following code:

```
from scipy.optimize import minimize_scalar

def steepest_descent(x_0, f, df, *f_params, df_eps = 1e-2, max_iter = 1000):
    iterations = []
    x = np.array(x_0)
    iterations.append(x)
    while np.linalg.norm(df(x, *f_params)) > df_eps and len(iterations) <= max_iter:
        res = minimize_scalar(lambda alpha: f(x - alpha * df(x, *f_params), *f_params))
        alpha_opt = res.x
        x = x - alpha_opt * df(x, *f_params)
        iterations.append(x)
    #print(f'Finished with {len(iterations)} iterations')
    return iterations

x_0 = [4,1]
k = 30
iterations = steepest_descent(x_0, f, df, k, df_eps = 1e-9)
title = f'$f(x_1, x_2) = x_1^2 + k x_2^2, k = {k}$'

plot_3d_function(x1, x2, f, title, k, minima=[0,0], iterations = iterations)
```
